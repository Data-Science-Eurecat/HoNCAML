```python
import textwrap
def wrap_labels(ax, width, break_long_words=False):
    labels = []
    for label in ax.get_yticklabels():
        text = label.get_text().replace('(','').replace(')','')
        text_split = text.split(',')[0]
        for i, config in enumerate(text.split(',')[1].split(' ')):
            if i > 0: text_split = text_split+'\n'+config
        labels.append(text_split)
    ax.set_yticklabels(labels, rotation=0)
```

```python
def base_viz(metric_name):
    base = alt.Chart(results_melt, title=metric_name).mark_bar(
    ).encode(
        x='value:Q',
        y=alt.Y('index_col:O', axis=alt.Axis(labels=False)),
        tooltip=['index_col_b','value'],
        color = alt.Color('index_col')
    ).properties(
        height=200,
        width=200,
    ).transform_filter(
        alt.datum.metric==metric_name
    )
    return base

row1 = alt.hconcat()
for metric_name in ['mean_squared_error','mean_absolute_percentage_error','median_absolute_error']:
    row1 |= base_viz(metric_name)
row2 = alt.hconcat()
for metric_name in ['r2_score','mean_absolute_error','root_mean_square_error']:
    row2 |= base_viz(metric_name)
row1 & row2
```

```python
# find the most recent execution folder
most_recent_execution = max(os.listdir('honcaml_reports'))
file_path = f'honcaml_reports/{most_recent_execution}/results.csv'
results = pd.read_csv(file_path)

results['model'] = results['config/model_module'].apply(lambda x: x.split('.')[-1])

results['configs'] = results.apply(
    lambda x: 'max_features:' + str(x['config/param_space/max_features']) + ' n_estimators:' +
              str(int(x['config/param_space/n_estimators']))
    if x['model'] == 'RandomForestRegressor'
    else 'fit_intercept:' + str(x['config/param_space/fit_intercept']), axis=1
)

results['model_configs'] = results.apply(
    lambda x: x['model'] + '<br>' + x['configs'].split(' ')[0] + '<br>' + x['configs'].split(' ')[1]
    if x['model'] == 'RandomForestRegressor'
    else x['model'] + '<br>' + x['configs'], axis=1
)

results = results[['model','configs','model_configs','mean_squared_error','mean_absolute_percentage_error',
                   'median_absolute_error','r2_score','mean_absolute_error','root_mean_square_error']]

results = results.drop_duplicates().reset_index().drop(columns=['index'])


results_melted = results[['model_configs','mean_squared_error','mean_absolute_percentage_error','median_absolute_error',
                        'r2_score','mean_absolute_error','root_mean_square_error']] \
    .melt(id_vars=['model_configs'], 
          value_vars=['mean_squared_error','mean_absolute_percentage_error', 'median_absolute_error','r2_score',
                  'mean_absolute_error','root_mean_square_error'], 
          var_name='metric')


fig = px.bar(results_melted, x="value", y="model_configs", color="model_configs", facet_col="metric", facet_col_wrap=3, 
             facet_row_spacing=0.1, opacity=0.)
fig.update_yaxes(visible=False)
fig.update_xaxes(title = None, matches=None, showticklabels=True)
fig.update_layout(legend=dict(title='Model & Configs'))
fig.for_each_annotation(lambda a: a.update(text=a.text.split("=")[-1]))
fig.show()

```
